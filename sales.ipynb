```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("sales.csv", encoding='unicode_escape')
```

```python
data["Amount"] = data["Amount"].fillna(data["Amount"].median())
```

```python
data["Amount"].isna().sum()
```

```python
data.drop(["Status","unnamed1"], axis=1)
```

```python
x = data["Gender"].unique()
x
```

```python
y = data["Amount"].sum()
y
```

```python
gender = data.groupby("Gender")["Amount"].sum()
gender
```

```python
a1 = gender.index
a2 = gender.values

sns.barplot(x=a1, y=a2, palette='PiYG', hue=x)

plt.title("Gen")

plt.show()
```

```python
age_group = data.groupby("Age Group")["Amount"].sum()
age_group

sns.barplot(x=age_group.index, y=age_group.values, palette="pastel", legend=False)

plt.title("Age Group", fontsize=20)
plt.xlabel("Age Group", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
marriage_status = data.groupby("Marital_Status")["Amount"].sum()
marriage_status

sns.barplot(x=marriage_status.index, y=marriage_status.values, palette="Dark2", hue=x, legend=False)

plt.title("Marriaged Couples ", fontsize=20)
plt.xlabel("Marriage Status", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
state = data.groupby("State")["Amount"].sum()
state


fig, ax = plt.subplots(figsize=(30, 5))

sns.lineplot(x=state.index, y=state.values, marker="o", ax=ax, markersize=20)

plt.title("States", fontsize=20)
plt.xlabel("States Name", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
zone = data.groupby("Zone")["Amount"].sum()
zone

sns.scatterplot(x=zone.index, y=zone.values, s=100)

plt.title("Zone", fontsize=20)
plt.xlabel("Zone Name", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
product_cat = data.groupby("Product_Category")["Amount"].sum()
product_cat

fig, ax = plt.subplots(figsize=(30,5))

sns.lineplot(x=product_cat.index, y=product_cat.values, ax=ax, marker="o", markersize=20, color="gray")

plt.title("Product Category", fontsize=20)
plt.xlabel("Product Name", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
order = data.groupby("Orders")["Amount"].sum()

color = [""]

sns.barplot(x=order.index, y=order.values, palette="BuPu")

plt.title("Order", fontsize=20)
plt.xlabel("Orders Name", fontsize=10)
plt.ylabel("Amount", fontsize=10)

plt.show()
```

```python
product_ord = data.groupby("Product_Category")["Orders"].sum()


fig, ax = plt.subplots(figsize=(30,7))

sns.barplot(x=product_ord.index, y=product_ord.values, palette="Blues_d", ax=ax)

plt.title("Product with Order Count", fontsize=20)
plt.xlabel("Product Name", fontsize=10)
plt.ylabel("Order", fontsize=10)

plt.show()
```
